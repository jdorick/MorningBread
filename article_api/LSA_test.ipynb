```python
from db_update.db_connector import *
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import DBSCAN
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
headlines_tuple = get_all_headlines()

headlines = []

for headline in headlines_tuple:
    headlines.append(headline[0])
```

```python
print(headlines)
```

```python
for x in headlines:
    print(x)
    
print(len(headlines))
```

```python
vectorizer = CountVectorizer(min_df=3, stop_words='english')
bag_of_words = vectorizer.fit_transform(headlines)
```

```python
print(bag_of_words.todense())
```

```python
svd = TruncatedSVD(n_components=2)
lsa = svd.fit_transform(bag_of_words)
```

```python
topic_encoded_df = pd.DataFrame(lsa, columns = ["topic_1", "topic_2"])
topic_encoded_df["headline"] = headlines
print(topic_encoded_df[["headline", "topic_1", "topic_2"]])
```

```python
dictionary = vectorizer.get_feature_names_out()
dictionary
```

```python
encoding_matrix = pd.DataFrame(svd.components_,
                               index=['topic_1', 'topic_2'],
                               columns=dictionary).T
encoding_matrix
```

```python
encoding_matrix['abs_topic_1'] = np.abs(encoding_matrix['topic_1'])
encoding_matrix['abs_topic_2'] = np.abs(encoding_matrix['topic_2'])
encoding_matrix.sort_values('abs_topic_1', ascending=False)
```

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(headlines)
```

```python
dbscan = DBSCAN(eps=1.0, min_samples=5)
dbscan.fit(vectors)
```

```python
cluster_labels = dbscan.labels_
coords = vectors.toarray()
```

```python
print(cluster_labels)
```

```python
no_clusters = len(np.unique(cluster_labels) )
no_noise = np.sum(np.array(cluster_labels) == -1, axis=0)

print('Estimated no. of clusters: %d' % no_clusters)
print('Estimated no. of noise points: %d' % no_noise)
```
